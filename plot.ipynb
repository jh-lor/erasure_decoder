```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "01242022_165234_10000.pkl"
df = pd.read_pickle(f"./df/{file}")
df
```

```python
num = sum(df.loc[(5,0.3), ["no_error", "undetected_error", "corrected_error", "uncorrected_error"]])
```

```python
plt.figure()
physical_error_rate = sorted(list(set(df.index.get_level_values("physical_error_rate"))))
plt.title(f"Logical Error Rate vs Physical Erasure Rate: {num:n} Simulations")
plt.xlabel("Physical Erasure Rate")
plt.ylabel("Logical Error Rate")
for size in sorted(list(set(df.index.get_level_values("size")))):
    plt.plot(physical_error_rate,df.loc[size, "logical_error_rate"], "--o", label = f"Size = {size}")
# plt.plot(physical_error_rate, df.loc[size,"effective_error_rate"], label = "Effective Error Rate")
plt.legend()
plt.savefig(f"{file}.jpg")
plt.show()
```
