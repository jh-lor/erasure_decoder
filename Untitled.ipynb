```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "12022021_144759_10000.pkl"
df = pd.read_pickle(f"./df/{file}")
df
```

```python
df.loc[5, "logical_error_rate"]
```

```python
plt.figure()
physical_error_rate = sorted(list(set(df.index.get_level_values("physical_error_rate"))))
plt.title("Logical Error Rate vs Physical Erasure Rate: 10^5 Simulations")
plt.xlabel("Physical Erasure Rate")
plt.ylabel("Logical Error Rate")
for size in sorted(list(set(df.index.get_level_values("size")))):
    plt.plot(physical_error_rate,df.loc[size, "logical_error_rate"], label = f"Size = {size}")
plt.plot(physical_error_rate, df.loc[size,"effective_error_rate"], label = "Effective Error Rate")
plt.legend()
plt.savefig("Rough_plot.jpg")
plt.show()
```
